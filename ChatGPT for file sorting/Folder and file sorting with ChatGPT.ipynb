```python
import os
import shutil

# Specify the base directory where the files are located
base_directory = "./Company Documents"

# Specify the file-to-folder mapping
file_to_folder_mapping = {
    "Board Meeting Minutes.txt": "Board Meetings",
    "Meeting Agendas.txt": "Board Meetings",
    "Meeting Presentations.txt": "Board Meetings",
    "Annual Financial Reports.txt": "Financial Reports",
    "Quarterly Financial Reports.txt": "Financial Reports",
    "Financial Projections.txt": "Financial Reports",
    "IT Security Policy.txt": "Policies and Guidelines",
    "Confidentiality Agreements.txt": "Policies and Guidelines",
    "Code of Conduct.txt": "Policies and Guidelines",
    "Data Privacy Policy.txt": "Policies and Guidelines",
    "Procurement Policy.txt": "Policies and Guidelines",
    "Workplace Safety Manual.txt": "Policies and Guidelines",
    "Conflict Resolution Policy.txt": "Policies and Guidelines",
    "Diversity and Inclusion Policy.txt": "Policies and Guidelines",
    "Performance Evaluations.txt": "Human Resources",
    "Employee Benefits Packages.txt": "Human Resources",
    "Job Descriptions.txt": "Human Resources",
    "Training Manuals.txt": "Human Resources",
    "Performance Improvement Plans.txt": "Human Resources",
    "Recruitment Campaigns.txt": "Human Resources",
    "Employee Onboarding Checklists.txt": "Human Resources",
    "Employee Recognition Programs.txt": "Human Resources",
    "Employee Exit Interviews.txt": "Human Resources",
    "Marketing Strategies.txt": "Marketing and Sales",
    "Sales Forecasts.txt": "Marketing and Sales",
    "Market Research Reports.txt": "Marketing and Sales",
    "Sales Contracts.txt": "Marketing and Sales",
    "Customer Satisfaction Surveys.txt": "Marketing and Sales",
    "Marketing Budgets.txt": "Marketing and Sales",
    "Marketing Campaigns.txt": "Marketing and Sales",
    "Social Media Content Calendars.txt": "Marketing and Sales",
    "Sales Training Manuals.txt": "Marketing and Sales",
    "Project Proposals.txt": "Projects and Proposals",
    "Project Status Updates.txt": "Projects and Proposals",
    "Project Charters.txt": "Projects and Proposals",
    "Risk Assessment Reports.txt": "Projects and Proposals",
    "Product Development Roadmaps.txt": "Projects and Proposals",
    "Supplier Contracts.txt": "Projects and Proposals",
    "Product Launch Plans.txt": "Projects and Proposals",
    "User Acceptance Testing Plans.txt": "Projects and Proposals",
    "Project Budgets.txt": "Projects and Proposals",
    "Project Timelines.txt": "Projects and Proposals",
    "Investor Presentations.txt": "Legal and Compliance",
    "Shareholder Agreements.txt": "Legal and Compliance",
    "Supplier Evaluation Scorecards.txt": "Legal and Compliance",
    "Service Level Agreements.txt": "Legal and Compliance",
    "Tax Compliance Guidelines.txt": "Legal and Compliance",
    "Corporate Governance Guidelines.txt": "Legal and Compliance",
    "IT Helpdesk Troubleshooting Guides.txt": "IT and Technology",
    "IT Infrastructure Upgrade Proposals.txt": "IT and Technology",
    "Data Backup and Recovery Plans.txt": "IT and Technology",
    "Inventory Management Systems.txt": "IT and Technology",
    "Software Release Documentation.txt": "IT and Technology",
    "Travel Expense Reports.txt": "Travel and Expenses",
    "Travel Authorization Forms.txt": "Travel and Expenses",
    "Conference Registrations.txt": "Travel and Expenses",
    "Travel Itineraries.txt": "Travel and Expenses",
    "Company Newsletters.txt": "Miscellaneous",
    "Corporate Social Responsibility Reports.txt": "Miscellaneous",
    "Performance Metrics Dashboards.txt": "Miscellaneous",
    "Expense Tracking Spreadsheets.txt": "Miscellaneous",
    "Corporate Branding Guidelines.txt": "Miscellaneous"
}

# Iterate over each file in the base directory
for file in os.listdir(base_directory):
    file_path = os.path.join(base_directory, file)
    if os.path.isfile(file_path):
        # Check if the file is in the mapping dictionary
        if file in file_to_folder_mapping:
            folder_name = file_to_folder_mapping[file]
            folder_path = os.path.join(base_directory, folder_name)
            # Create the folder if it doesn't exist
            if not os.path.exists(folder_path):
                os.makedirs(folder_path)
            # Move the file to the corresponding folder
            source_path = os.path.join(base_directory, file)
            destination_path = os.path.join(folder_path, file)
            shutil.move(source_path, destination_path)
            print(f"Moved file: {file} to folder: {folder_name}")

print("All files have been sorted into their respective folders.")
```

```python
import os

# Specify the base directory where the files are located
base_directory = "./Company Documents"

# Define the expected folder structure
expected_folder_structure = {
    "Board Meetings": ["Board Meeting Minutes.txt", "Meeting Agendas.txt", "Meeting Presentations.txt"],
    "Financial Reports": ["Annual Financial Reports.txt", "Quarterly Financial Reports.txt", "Financial Projections.txt"],
    "Policies and Guidelines": ["Employee Handbook.txt", "IT Security Policy.txt", "Confidentiality Agreements.txt",
                                "Code of Conduct.txt", "Data Privacy Policy.txt", "Procurement Policy.txt",
                                "Workplace Safety Manual.txt", "Conflict Resolution Policy.txt",
                                "Diversity and Inclusion Policy.txt"],
    "Human Resources": ["Performance Evaluations.txt", "Employee Benefits Packages.txt", "Job Descriptions.txt",
                        "Training Manuals.txt", "Performance Improvement Plans.txt", "Recruitment Campaigns.txt",
                        "Employee Onboarding Checklists.txt", "Employee Recognition Programs.txt",
                        "Employee Exit Interviews.txt"],
    "Marketing and Sales": ["Marketing Strategies.txt", "Sales Forecasts.txt", "Market Research Reports.txt",
                            "Sales Contracts.txt", "Customer Satisfaction Surveys.txt", "Marketing Budgets.txt",
                            "Marketing Campaigns.txt", "Social Media Content Calendars.txt",
                            "Sales Training Manuals.txt"],
    "Projects and Proposals": ["Project Proposals.txt", "Project Status Updates.txt", "Project Charters.txt",
                               "Risk Assessment Reports.txt", "Product Development Roadmaps.txt",
                               "Supplier Contracts.txt", "Product Launch Plans.txt", "User Acceptance Testing Plans.txt",
                               "Project Budgets.txt", "Project Timelines.txt"],
    "Legal and Compliance": ["Investor Presentations.txt", "Shareholder Agreements.txt",
                             "Supplier Evaluation Scorecards.txt", "Service Level Agreements.txt",
                             "Tax Compliance Guidelines.txt", "Corporate Governance Guidelines.txt"],
    "IT and Technology": ["IT Helpdesk Troubleshooting Guides.txt", "IT Infrastructure Upgrade Proposals.txt",
                          "Data Backup and Recovery Plans.txt", "Inventory Management Systems.txt",
                          "Software Release Documentation.txt"],
    "Travel and Expenses": ["Travel Expense Reports.txt", "Travel Authorization Forms.txt",
                            "Conference Registrations.txt", "Travel Itineraries.txt"],
    "Miscellaneous": ["Company Newsletters.txt", "Corporate Social Responsibility Reports.txt",
                      "Performance Metrics Dashboards.txt", "Expense Tracking Spreadsheets.txt",
                      "Corporate Branding Guidelines.txt"]
}

# Initialize counters for correctly sorted files and false positives
correctly_sorted_files = 0
false_positives = 0

# Iterate over each folder in the expected folder structure
for folder, files in expected_folder_structure.items():
    folder_path = os.path.join(base_directory, folder)
    # Iterate over each file in the folder
    for file in files:
        file_path = os.path.join(folder_path, file)
        # Check if the file exists in the expected folder
        if os.path.isfile(file_path):
            correctly_sorted_files += 1
        else:
            false_positives += 1

# Calculate the total number of files
total_files = correctly_sorted_files + false_positives

# Print the benchmarking results
print("Benchmarking Results:")
print("---------------------")
print(f"Total Files: {total_files}")
print(f"Correctly Sorted Files: {correctly_sorted_files}")
print(f"False Positives: {false_positives}")
```
